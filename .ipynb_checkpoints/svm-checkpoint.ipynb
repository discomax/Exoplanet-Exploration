```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from keras.utils import to_categorical
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("cumulative.csv")
df = df.drop(columns=["rowid", "kepid", "kepoi_name", "kepler_name", "koi_pdisposition", "koi_score", "koi_tce_delivname"])
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

```python
# Encode the koi_disposition column (will be the target)

target_names = ["CANIDATE", "CONFIRMED", "FALSE POSITIVE"]

label_encoder = LabelEncoder()
df['koi_disposition'] = label_encoder.fit_transform(df['koi_disposition'])
df.tail()
# encoded_target = label_encoder.transform(target).reshape(-1, 1)
# encoded_target.shape
```

```python
data = df.drop("koi_disposition", axis=1)
feature_names = data.columns

target = df["koi_disposition"].values.reshape(-1,1)

print(data.shape, target.shape)
target
```

# Create a Train Test Split

Use `koi_disposition (one_hot_target)` for the y values

```python
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)

X_train
```

# Pre-processing

Scale the data using the MinMaxScaler

```python
# Create a MinMaxScaler model and fit it to the training data
X_scaler = MinMaxScaler().fit(X_train)
y_scaler = MinMaxScaler().fit(y_train)

# Transform the train and test data w/ the X and y scaler
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
y_train_scaled = y_scaler.transform(y_train)
y_test_scaled = y_scaler.transform(y_test)

y_train_scaled
```

# Train the Support Vector Machine

```python
from sklearn.svm import SVC

model = SVC(kernel = 'linear')
#model.fit(X_train_scaled, y_train_scaled)
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the `C` and `gamma` parameters

```python
from sklearn.model_selection import GridSearchCV

# Create the GridSearchCV estimator
param_grid = {'C': [1, 10, 100],
              'gamma': [0.0001, 0.001, 0.01]}
grid = GridSearchCV(model, param_grid, verbose=3)
```

```python
# Train the model with GridSearch
grid.fit(X_train_scaled, y_train.ravel())
```

```python
print(grid.best_params_)
print(grid.best_score_)
```
